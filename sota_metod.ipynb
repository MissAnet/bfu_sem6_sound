```python
import os
from pathlib import Path
from clearvoice import ClearVoice

# 1. Инициализация модели
# MossFormer2_SE_48K обеспечивает отличное качество для всех типов файлов
myClearVoice = ClearVoice(
    task='speech_enhancement', 
    model_names=['MossFormer2_SE_48K']
)

# 2. Настройка путей
input_folder = Path('dataset')
output_folder = Path('cleaned_results_sota')

# Создаем папку для результатов, если её нет
output_folder.mkdir(parents=True, exist_ok=True)

# 3. Список поддерживаемых расширений
extensions = {'.wav', '.mp3', '.ogg', '.flac', '.m4a'}

print(f"Начинаю обработку файлов из папки: {input_folder}")

# 4. Цикл обработки
# rglob('*') найдет файлы даже во вложенных папках (если нужно только в корне, используйте glob('*'))
files_to_process = [f for f in input_folder.glob('*') if f.suffix.lower() in extensions]

if not files_to_process:
    print("Подходящие аудиофайлы не найдены.")
else:
    for file_path in files_to_process:
        print(f"Обработка: {file_path.name}...")
        
        try:
            # Путь для сохранения (меняем только папку, имя оставляем оригинальное)
            save_path = output_folder / f"enhanced_{file_path.stem}.wav"
            
            # Обработка
            # ClearVoice внутри использует librosa/soundfile, поэтому понимает разные форматы
            output_wav = myClearVoice(input_path=str(file_path), online_write=False)
            
            # Сохранение (обычно ClearVoice сохраняет в .wav для максимального качества)
            myClearVoice.write(output_wav, output_path=str(save_path))
            
        except Exception as e:
            print(f"Ошибка при обработке файла {file_path.name}: {e}")

print(f"\nГотово! Все очищенные файлы находятся в папке: {output_folder}")
```
